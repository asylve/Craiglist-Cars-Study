```python
import pandas as pd
cars = pd.read_csv('cars_cleaned.csv')
cars.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in ['condition', 'drive', 'fuel', 'location', 'paint color', 'sale type', 'title status', 'transmission', 'type', 
            'cylinders', 'size', 'year']:
    dat = cars[col].value_counts()
    print(col)
    sns.barplot(x = dat.index, y = dat, figsize=(15,8))
    plt.show()
```

```python
#the civic is by far the most popular car on craiglist in the area
cars.model0.value_counts()[:50].plot.bar(figsize=(15,8))
```

```python
com_cars = cars.model0.value_counts()[:50].index

com_price = cars[cars.model0.isin(com_cars)].groupby(['model0']).price.mean()
com_price.plot.bar(figsize=(15,8))
```
